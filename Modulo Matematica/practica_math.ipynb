```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sympy as sp

# Define la variable simbólica y la función matemática
x = sp.symbols('x')
funcion = x**2 + 2*x + 1

# Convierte la función Sympy en una función NumPy para poder evaluarla en un rango de valores
funcion_numpy = sp.lambdify(x, funcion, 'numpy')

# Crea un rango de valores para x
x_vals = np.linspace(-5, 5, 100)

# Evalúa la función en el rango de valores de x
y_vals = funcion_numpy(x_vals)

# Configura el estilo de Seaborn
sns.set(style="whitegrid")

# Crea el gráfico con Matplotlib
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='Función: $x^2 + 2x + 1$')
plt.title('Gráfico de una función matemática')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
